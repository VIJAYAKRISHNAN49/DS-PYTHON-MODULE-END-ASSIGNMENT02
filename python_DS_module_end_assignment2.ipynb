```python
import statistics as stat
```

```python
Qmean=np.mean(df['Quantity'])
Qmed=np.median(df['Quantity'])
Qmod=stat.mode(df['Quantity'])
Qstd=stat.stdev(df['Quantity'])
Qvar=stat.variance(df['Quantity'])
Qper = np.percentile(df['Quantity'], 75)
print('mean,median and mode of qunatity:', Qmean,Qmed,Qmod)
print('std,var and percentile of qunatity:', Qstd,Qvar,Qper)
```

```python

Ustd=stat.stdev(df['UnitPrice'])
Uvar=stat.variance(df['UnitPrice'])
Uper=np.percentile(df['UnitPrice'],25)
#print('mean,median and mode of qunatity:', Qmean,Qmed,Qmod)
print('std,var and percentile of qunatity:', Ustd,Uvar,Uper)
```

```python
Tmean=np.mean(df['TotalPrice'])
Tmed=np.median(df['TotalPrice'])
Tmod=stat.mode(df['TotalPrice'])
print('mean,median and mode of TotalPrice:', Tmean,Tmed,Tmod)
```

```python
df['Month']=df['InvoiceDate'].dt.month

monthly_sales = df.groupby('Month')['TotalPrice'].sum()
monthly_sales.index = monthly_sales.index.astype(str)
# Plot
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot
plt.figure(figsize=(12,6))
plt.bar(monthly_sales.index, monthly_sales.values)
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot
plt.figure(figsize=(10,6))
plt.hist(df['TotalPrice'], bins=30)

plt.title('Distribution of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['TotalPrice'])

plt.title('Box Plot of Total Price')
plt.ylabel('Total Price')
plt.grid(True)
```

```python
plt.figure(figsize=(12,6))

top_countries = df['Country'].value_counts().head(10).index

sns.countplot(
    data=df[df['Country'].isin(top_countries)],
    x='Country',
    order=top_countries
)

plt.title('Count of Transactions by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(5).index

plt.figure(figsize=(12,6))

sns.violinplot(
    data=df[df['Country'].isin(top_countries)],
    x='Country',
    y='TotalPrice'
)

plt.title('Total Price Distribution by Country')
plt.show()
```

```python
numeric_cols = ['Quantity', 'UnitPrice', 'TotalPrice']

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()
```

```python
country_sales = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)

print(country_sales.head())

top_country = country_sales.idxmax()
print("Top Country:", top_country)
country_sales.head(10).plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Countries by Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Month'] = df['InvoiceDate'].dt.month_name()

monthly_sales = df.groupby('Month')['TotalPrice'].sum()

print(monthly_sales.sort_values(ascending=False))
monthly_sales.plot(kind='bar', figsize=(10,5))
plt.title('Sales by Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Hour'] = df['InvoiceDate'].dt.hour

hourly_sales = df.groupby('Hour')['TotalPrice'].sum()

print(hourly_sales.sort_values(ascending=False))
hourly_sales.plot(kind='line', marker='o', figsize=(10,5))

plt.title('Sales by Hour')
plt.xlabel('Hour')
plt.ylabel('Total Sales')
plt.grid(True)

plt.show()
```
